```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_bgr = cv2.imread('./zumi0.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(img_bgr)
plt.subplot(122)
plt.imshow(img_rgb)
```

```python
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(img_rgb)
plt.subplot(122)
plt.imshow(img_hsv)
```

```python
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.cornerHarris(img_gray, 2, 3, 0.04)
```

```python
plt.imshow(corners, cmap='gray')
```

```python
#This is important
import cv2 as cv
```

```python
sift = cv.xfeatures2d.SIFT_create()
```

```python
kp = sift.detect(img_bgr)
```

```python
import numpy as np
img_kp = np.zeros_like(img_bgr)
img_kp = cv.drawKeypoints(img_rgb,kp,img_kp)
```
